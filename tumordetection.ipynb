```python
import os
import cv2

# Define the categories
categories=['small_Benign','small_Malignant']

data = []

# Loop through each category
for category in categories:
    path = os.path.join('train', category)  # Assuming the folders are in the root of your Google Drive
    label = categories.index(category)

    # Loop through each file in the category folder
    for file in os.listdir(path):
        img_path = os.path.join(path, file)

        # Read the image
        img = cv2.imread(img_path)

        # Resize the image to 224x224
        try:
            img = cv2.resize(img, (224, 224))
        except Exception as e:
            print(f"Error processing {img_path}: {str(e)}")
            continue

        # Append the image and its label to the data list
        data.append([img, label])

# Now you have a list 'data' containing image-label pairs
# Each element of 'data' is a list [image, label], where 'image' is a numpy array and 'label' is an integer
# You can further process this data for your machine learning model
```

```python
from keras.applications.vgg16 import VGG16
from keras import Sequential

# Load the pre-trained VGG16 model
vgg = VGG16()

# Create a new Sequential model
model = Sequential()

# Add layers from VGG16 to the new model, ignoring the last layer
for layer in vgg.layers[:-1]:  # Ignore the last layer
    model.add(layer)

# Display the summary of the new model
model.summary()
#see all same but last one has ignore
```

```python
#we have train the parameter
#now freeze all the layers
#ab hum sare ke sare layers ko freeze karage
#so train karne par weight update ho
#now we have freeze all layers( because weight and bais not update)
for layer in model.layers:
    layer.trainable=False
```

```python
#ye karne ke baat sab non-trainable band jayegey
#sab karne par aisa dekhaya dega
#ab saare trainable =0 hai, jitne bhi parameter hai vo non-trianable hai
#as a freeze ho gay ab update nahi kar sakte as a training time
model.summary()
```

```python
#abb hamara wal work karna hai ,hamara wala last layer main add hoga
from keras.layers import Dense
```

```python
model.add(Dense(units=1,activation='sigmoid'))
```

```python
model.compile(optimizer='Adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
trained_model= model.fit(x_train,y_train,epochs=5,validation_data=(x_test,y_test))
```
